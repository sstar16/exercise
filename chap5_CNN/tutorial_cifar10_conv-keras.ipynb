## 准备数据

```python
import os
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, optimizers, datasets
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D

os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'  # or any {'0', '1', '2'}

def cifar10_dataset():
    (x, y), (x_test, y_test) = datasets.cifar10.load_data()
    ds = tf.data.Dataset.from_tensor_slices((x, y))
    ds = ds.map(prepare_mnist_features_and_labels)
    ds = ds.take(20000).shuffle(20000).batch(100)
    
    test_ds = tf.data.Dataset.from_tensor_slices((x_test, y_test))
    test_ds = test_ds.map(prepare_mnist_features_and_labels)
    test_ds = test_ds.take(20000).batch(20000)
    return ds, test_ds

def prepare_mnist_features_and_labels(x, y):
    x = tf.cast(x, tf.float32) / 255.0
    y = tf.cast(y, tf.int64)
    return x, y
```

## 建立模型

```python
class myConvModel(keras.Model):
    '''在这里实现alexNet模型'''
    def __init__(self):
        super(myConvModel, self).__init__()
        self.conv1 = Conv2D(64, kernel_size=(3, 3), padding='same', activation='relu')
        self.conv2 = Conv2D(64, kernel_size=(3, 3), padding='same', activation='relu')
        self.pool1 = MaxPooling2D(pool_size=(2, 2), strides=2)
        
        self.conv3 = Conv2D(128, kernel_size=(3, 3), padding='same', activation='relu')
        self.conv4 = Conv2D(128, kernel_size=(3, 3), padding='same', activation='relu')
        self.pool2 = MaxPooling2D(pool_size=(2, 2), strides=2)
        
        self.conv5 = Conv2D(256, kernel_size=(3, 3), padding='same', activation='relu')
        self.conv6 = Conv2D(256, kernel_size=(3, 3), padding='same', activation='relu')
        self.conv7 = Conv2D(256, kernel_size=(3, 3), padding='same', activation='relu')
        self.pool3 = MaxPooling2D(pool_size=(2, 2), strides=2)
        
        self.flat = Flatten()
        self.dense1 = Dense(1024, activation='relu')
        self.dropout1 = Dropout(0.5)
        self.dense2 = Dense(1024, activation='relu')
        self.dropout2 = Dropout(0.5)
        self.dense3 = Dense(10)
    
    @tf.function
    def call(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.pool1(x)
        
        x = self.conv3(x)
        x = self.conv4(x)
        x = self.pool2(x)
        
        x = self.conv5(x)
        x = self.conv6(x)
        x = self.conv7(x)
        x = self.pool3(x)
        
        x = self.flat(x)
        x = self.dense1(x)
        x = self.dropout1(x)
        x = self.dense2(x)
        x = self.dropout2(x)
        logits = self.dense3(x)
        return logits
    
    @tf.function
    def getL1_feature_map(self, x):
        x = self.conv1(x)
        return x
    
    @tf.function
    def getL2_feature_map(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.pool1(x)
        x = self.conv3(x)
        return x

model = myConvModel()
optimizer = optimizers.Adam(0.001)
```

```python
# 调试：检查数据管道
train_ds, test_ds = cifar10_dataset()

# 检查一个批次的数据
for batch_x, batch_y in train_ds.take(1):
    print(f"Batch x shape: {batch_x.shape}")  # 应该是(100, 32, 32, 3)
    print(f"Batch y shape: {batch_y.shape}")  # 应该是(100,)
    print(f"Batch y dtype: {batch_y.dtype}")  # 应该是int64
    print(f"Batch y values sample: {batch_y[:5]}")  # 应该是[0-9]的整数
    break

# 测试模型前向传播
model = myConvModel()
dummy_input = tf.random.normal([1, 32, 32, 3])
output = model(dummy_input)
print(f"Model output shape: {output.shape}")  # 应该是(1, 10)
print(f"Model output: {output}")
```

## 编译， fit以及evaluate

```python
model.compile(optimizer=optimizer,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
train_ds, test_ds = cifar10_dataset()
model.fit(train_ds, epochs=10)
model.evaluate(test_ds)
```
